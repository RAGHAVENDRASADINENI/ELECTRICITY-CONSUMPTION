# MODEL BUILDING

```python
y=train_clean.electricity_consumption
X=train_clean.drop("electricity_consumption",axis=1)
```

```python
y=y.astype("int64")
```

```python
y.plot(kind="box",vert=False)
```

```python
np.log(y).plot(kind="box",vert=False)
```

```python
np.sqrt(y).plot(kind="box",vert=False)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree=DecisionTreeRegressor(max_depth=12)
```

```python
treemodel=tree.fit(X,y)
```

```python
treemodel.score(X,y)
```

```python
treepredict=treemodel.predict(X)
```

```python
treeresid=y-treepredict
```

```python
np.sqrt(np.mean(treeresid**2))
```
